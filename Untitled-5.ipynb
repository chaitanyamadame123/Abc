```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3])
y = np.array([1, 5, 2])

coefficients = np.polyfit(x, y, 2)
x_plot = np.linspace(1, 3.5, 100)
y_plot = np.polyval(coefficients, x_plot)
plt.plot(x_plot, y_plot, label='Interpolating Polynomial')
plt.scatter(x, y, color='red', label='Points')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.title('Interpolating Polynomial')
plt.legend()
plt.grid(True)
plt.show()
```
