```python
def cube_root_if_break(n):
    guess = 1.0
    while True:
        if abs(guess**3 - n) < 0.0001:
            break
        guess = (2*guess + n/(guess**2))/3
    return guess

# Method 2: Using iteration counter
def cube_root_iteration_counter(n):
    guess = 1.0
    iterations = 0
    while True:
        if abs(guess**3 - n) < 0.0001:
            break
        guess = (2*guess + n/(guess**2))/3
        iterations += 1
        if iterations > 1000:
            break
    return guess

# Method 3: Using For Loop on a guess value
def cube_root_for_loop(n):
    for guess in range(1, 1001):
        if abs(guess**3 - n) < 0.0001:
            return guess
    return None

# Test the functions
cube_root1 = cube_root_if_break(27)
cube_root2 = cube_root_iteration_counter(27)
cube_root3 = cube_root_for_loop(27)

print("Cube root of 27 using if-break statement:", cube_root1)
print("Cube root of 27 using iteration counter:", cube_root2)
print("Cube root of 27 using for loop on guess value:", cube_root3)
```

```python
import matplotlib.pyplot as plt

def plot_guess_and_error(n):
    guesses = []
    errors = []
    for guess in range(1, 101):
        guesses.append(guess)
        errors.append(guess**3 - n)

    plt.plot(guesses, errors)
    plt.xlabel('Guess')
    plt.ylabel('Error (guess^3 - cube)')
    plt.title('Guess Value vs Error')
    plt.show()

# Test the function
plot_guess_and_error(27)
```
